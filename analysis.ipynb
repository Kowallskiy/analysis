```python
climate_hist['DATE'] = pd.to_datetime(climate_hist['DATE'])
```

```python
climate_hist.info()
```

```python
climate_hist['month'] = climate_hist['DATE'].dt.month
climate_hist['year'] = climate_hist['DATE'].dt.year
```

```python
climate_hist.head()
```

```python
average_temp_each_month = climate_hist.groupby(['year', 'month']).mean()['TAVG']
```

```python
average_temp_each_month = pd.DataFrame(average_temp_each_month)
average_temp_each_month.reset_index(inplace=True)
average_temp_each_month.head()
```

```python
month_name = {
    1: "Jan",
    2: "Feb",
    3: "Mar",
    4: "Apr",
    5: "May",
    6: "Jun",
    7: "Jul",
    8: "Aug",
    9: "Sep",
    10: "Oct",
    11: "Mov",
    12: "Dec"
}

average_temp_each_month['date'] = pd.Series(f"{y}-{month_name[m]}" for y, m in zip(average_temp_each_month['year'], average_temp_each_month['month']))
```

```python
plt.figure(figsize=(14, 8))
plt.title("Average Temperature")

plt.plot(average_temp_each_month['date'], average_temp_each_month['TAVG'], label='average temperature')
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.xticks(rotation=90)
plt.show()
```

```python
average_temp_year = average_temp_each_month.groupby("month").mean('TAVG')
```

```python
average_temp_year.head()
```

```python
season_name = {
    1: "Winter",
    2: "Winter",
    3: "Spring",
    4: "Spring",
    5: "Spring",
    6: "Summer",
    7: "Summer",
    8: "Summer",
    9: "Fall",
    10: "Fall",
    11: "Fall",
    12: "Winter"
}

average_temp_year = average_temp_year.reset_index()
average_temp_year.drop(columns=['year'], inplace=True)
average_temp_year['season'] = pd.Series(season_name[s] for s in average_temp_year['month'])
```

```python
average_temp_year.head()
```

```python
plt.figure(figsize=(14, 8))

plt.title("Average temperature each month")
plt.plot(average_temp_year.month, average_temp_year['TAVG'])
plt.ylabel("Temperature")
plt.xlabel("Month")
plt.grid(True)

plt.show()
```

```python
from matplotlib_venn import venn2

hist_pest_and_dis_crops = set(hist_pest_and_dis["Crop_Type"].unique())
pest_df_crops = set(pest_df["Crop for which pest data is being collected"].unique())

# Create the Venn diagram
venn = venn2([hist_pest_and_dis_crops, pest_df_crops], ('History Pest and Disease Data', 'Pest Data'))

# Customize the plot
plt.title("Comparison of Crops Between Two Datasets")
plt.show()
```

```python
hist_pest_and_dis.head()
```

```python
def safe_to_datetime(date_str):
    try:
        return pd.to_datetime(date_str, format="%m/%d/%Y")
    except ValueError as e:
        print(f"Error! The date is {date_str}")
        return None
        
hist_pest_and_dis['Date of Observation'] = hist_pest_and_dis['Date of Observation'].apply(safe_to_datetime)
```

```python
hist_pest_and_dis.info()
```

```python
hist_pest_and_dis.dropna(inplace=True)
```

```python
hist_pest_and_dis.sort_values(by="Date of Observation", ascending=True, inplace=True)
```

```python
hist_pest_and_dis.head()
```

```python
hist_pest_and_dis.nunique()
```

```python
hist_pest_and_dis['month'] = hist_pest_and_dis['Date of Observation'].dt.month
hist_pest_and_dis = hist_pest_and_dis.drop(columns='Date of Observation')
hist_pest_and_dis.head(10)
```

```python
crop_month_count = hist_pest_and_dis.groupby(['month', "Crop_Type"]).size().reset_index(name='count')
```

```python
crop_month_count.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for crop in crop_month_count['Crop_Type'].unique():
    crop_type = crop_month_count[crop_month_count['Crop_Type'] == crop]
    ax.plot(crop_type['month'], crop_type['count'], marker='o', label=crop)

ax.set_title("Number of pest/disease occurences each month")
ax.set_xlabel("Month")
ax.set_ylabel("Count")
ax.legend(title="Crop Type")
ax.grid(True)

plt.xticks(crop_month_count['month'].unique())
plt.show()
```

## Example

```python
def get_recommendations_pest(plant, predicted_pest):
    pest_data = pest_df[(pest_df["Crop for which pest data is being collected"] == plant) & (pest_df["Common_Name"] == predicted_pest)]
    # print(pest_data.empty)
    if not pest_data.empty:
        row = pest_data.iloc[0]
        recommendations = {
            "Plant": plant,
            "Pest": predicted_pest,
            "Seasonal Appearance": row["Seasonal_Appearance"],
            "Feeding Habit": row['Feeding_Habit'],
            "Feeding Target": row['Feeding_Target'],
            "Damage Patterns": row['Damage_Patterns'],
            "Control Susceptibility": row['Control_Susceptibility'],
            "Crop Susceptibility Stage": row['Crop_Susceptibility_Stage'],
            "Crop Preventive Measures": row['Crop_Preventive_Measures'],
            "Non-pesticidal control measures": row['Non-pesticidal_control_measures'],
            "Chemical control measures": row['Chemical_control_measures'],
        }
        return recommendations
    else:
        recommendations = {
            "Sorry": "We have no idea what is going on with your plant. Go to a plant doctor"
        }
        return recommendations
```

```python
rec = get_recommendations_pest("Kale", "Cabbage Whitefly")

for key, val in rec.items():
    print(f"{key}: {val}")
```
